solve 
$$
z=5x_1+4x_2\\
subject,\\
6x_1+4x_2\le24\\
x_1+2x_2\le6\\
x_2-x_1\le1\\
x_2\le2\\
x_1,x_2\ge0\\
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sympy import *
```

# constraints

```python
f_1=lambda x:(24-6*x)/4
f_2=lambda x:(6-x)/2
f_3=lambda x:1+x
f_4=lambda x,y:5*x+4*y
```

# objective function

```python
x,y=symbols('x y',real=True)
e1=Eq((6*x+4*y-24),0)
e2=Eq((x+2*y-6),0)
e3=Eq((y-x-1),0)
e4=Eq((5*x+4*y),0)
```

```python
d=[]

#solution for e1,e2
sol1=(solve((e1,e2),(x,y)))
d.append(sol1[x])
d.append(sol1[y])
sol1=d
#solution for e4,e3
d=[]
sol2=(solve((e1,e3),(x,y)))
d.append(sol2[x])
d.append(sol2[y])
sol2=d
#solution for e1,e3
d=[]
sol3=(solve((e1,e4),(x,y)))
d.append(sol3[x])
d.append(sol3[y])
sol3=d
#solution for e2,e3
d=[]
sol4=(solve((e2,e3),(x,y)))
d.append(sol4[x])
d.append(sol4[y])
sol4=d
#solution for e2,e4
d=[]
sol5=(solve((e2,e4),(x,y)))
d.append(sol5[x])
d.append(sol5[y])
sol5=d
print(sol5)
#solution for e3,e4
d=[]
sol6=(solve((e3,e4),(x,y)))
d.append(sol6[x])
d.append(sol6[y])
sol6=d
sol6
```

```python
maximum=0
point=[]

maximum_new=f_4(sol1[0],sol1[1])
if maximum_new>maximum:
    maximum=maximum_new
    point=sol1

maximum_new=f_4(sol2[0],sol2[1])
if maximum_new>maximum:
    maximum=maximum_new
    point=sol2

maximum_new=f_4(sol3[0],sol3[1])
if maximum_new>maximum:
    maximum=maximum_new
    point=sol3
    
maximum_new=f_4(sol4[0],sol4[1])
if maximum_new>maximum:
    maximum=maximum_new
    point=sol4
    
maximum_new=f_4(sol5[0],sol5[1])
if maximum_new>maximum:
    maximum=maximum_new
    point=sol5
    
maximum_new=f_4(sol6[0],sol6[1])
if maximum_new>maximum:
    maximum=maximum_new
    point=sol6
    
print(maximum)
print(f"maximum at point {point} ")
```

```python
X=np.linspace(-2,10,100)
z=[]
a=[]
b=[]
x=[]
c=[]
for i in range(len(X)):
    z.append(f_1(X[i]))
    a.append(f_2(X[i]))
    b.append(f_3(X[i]))
for i in range(len(X)):
    x.append(2)
plt.figure(figsize=(10,8))
plt.plot()
plt.xlim(-4,10)
plt.ylim(-4,10)
plt.title("solving the lpp in graphical method")
plt.axvline(x=0,color="black")
plt.axhline(y=0,color="black")
plt.plot(z,X,label="$6x_1+4x_2-24<=0$")
plt.plot(a,X,label="$x_1+2x_2<= 0$")
plt.plot(X,b,label="$x_2-x_1<=1$")
plt.plot(X,x,label="$x_2<=2$")
plt.scatter(0,0)
plt.scatter(2,2)
plt.scatter(1,2)
plt.scatter(0,1)
plt.legend(loc=1)
plt.show()
```
