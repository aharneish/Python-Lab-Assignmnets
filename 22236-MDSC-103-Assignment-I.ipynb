# Golden Section
formula
$$x_2=x_L+\delta (x_R-x_L)
x_1=x_R-\delta (x_R-x_L) $$
conditions

$f(x_1)>f(x_2)$ then $I_i=[x_L,x_2]$
$f(x_1)<f(x_2)$ then $I_i=[x_1,x_R]$
$f(x_1)=f(x_2)$ then $I_i=[x_1,x_2]$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=lambda x: x**2
y=[]
X=np.linspace(-100,100,100)
for i in range(len(X)):
    y.append(f(X[i]))
#plt.figure(figsize=(20,10))
plt.scatter(X,y)
plt.show()
xl=float(input("enter the starting point for the left: "))
xr=float(input("enter the starting point for the right: "))
iter=int(input("for how many iterations do you want to do: "))
i=0
x1=0
x2=0
#$$x_2=x_L+\delta (x_R-x_L)
#x_1=x_R-\delta (x_R-x_L) 
while i<=iter:
    x1=xr-0.618*(xr-xl)
    x2=xl+0.618*(xr-xl)
    if f(x1)>f(x2):
        xr=x2
    elif f(x1)<f(x2):
        xl=x1
    elif f(x1)==f(x2):
        xl=x1
        xr=x2
    print(f"for iteration {i} the values for xl,xr is [{x1},{x2}]")
    plt.plot(xl,)
    i=i+1
print(f"the points after iteration {iter} is {x1},{x2}")
```

# netwon raphson method
$$f(x)=x_1 ^ 3 + x_2 ^ 3-3 x_1 x_2$$


```python
from sympy import symbols, diff
from sympy.matrices import Matrix, Inverse, Transpose
import numpy as np
def grad_f():
    x, y = symbols('x y', real = True)
    f = x**3 + y**3 - 3*x*y
    grad_f_trans = [diff(f,x),diff(f,y)]
    return grad_f_trans
def jacobi():        #takes grad_f_trans as an input
    gf = grad_f()
    x, y = symbols('x y', real = True)
    j_matrix = [[diff(gf[0],x), diff(gf[0],y)],[diff(gf[1],x), diff(gf[1],y)]]
    return j_matrix



def N_R_Method():    #takes j_matrix as an input
    x0 = Matrix([0,0])
    gf_trans = grad_f()
    jacobi_M = jacobi()
    x, y = symbols('x y', real = True)
    jacobi_M = [[jacobi_M[0][0].subs(x, x0[0]).subs(y, x0[1]), jacobi_M[0][1].subs(x,x0[0]).subs(y,x0[1])],[jacobi_M[1][0].subs(x,x0[0]).subs(y,x0[1]),jacobi_M[1][1].subs(x,x0[0]).subs(y,x0[1])]]
    jacobi_M = Matrix(jacobi_M)
    jacobi_M_inv = Inverse(jacobi_M)
    g_x0 = [gf_trans[0].subs(x,x0[0]).subs(y,x0[1]), gf_trans[1].subs(x,x0[0]).subs(y,x0[1])]
    g_x0 = Matrix(g_x0)
    for i in range(0,50):
        x1 = x0 - jacobi_M_inv * g_x0
        x0 = x1
    return x0
N_R_Method()
```

# gradient desent
$$X_k+1 = X_k +r_k \delta f(X_k)$$
$r_k$ i s  fixed to 0.2 

example question
$$Maximize f(x_1,x_2)=4 x_1 + 6 x_2-2 x_1 ^2 - 2x_1 x_2- 2 x_2 ^2$$
let the starting point $X_0$ be (1,1)

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
X=np.linspace(-10,10,100)

x,y=symbols('x y',real=True)
f=lambda x: x**4 
z=[]
for i in range(len(X)):
    z.append(f(X[i]))
plt.plot(z,label="F(x)")
plt.legend()
plt.show()
def gradf():
    x=symbols('x',real=True)
    f= x**4
    gradf_trans=diff(f,x)
    return gradf_trans
def gradient_descent():
    #Xk+1 = Xk + rk∇f1Xk2
    x0=1
    x=symbols('x',real=True)
    r=0.02 #learning rate is fixed here
    gf=gradf()
    g_x0=gf.subs(x,x0)
    for i in range(0,30):
        x1=x0-r*g_x0
        x0=x1
    return x0
gradient_descent()
```
